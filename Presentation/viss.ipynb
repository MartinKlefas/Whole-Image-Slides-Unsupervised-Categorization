```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 2
means_frank = (220025, 57458)
means_guido = (4200, 800)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.15
opacity = 0.8

rects1 = plt.bar(index, means_frank, bar_width, alpha=opacity, color='b',label='Kaggle')

rects2 = plt.bar(index + bar_width, means_guido, bar_width, alpha=opacity, color='g', label='UMCM')

plt.xlabel('Dataset')
plt.ylabel('Number of tiles')
plt.title('Datasets breakdown')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D'))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# plt.figure(figsize=(18, 12))
# data to plot
n_groups = 6
training = (0.50, 0.59, 0.63, 0.61, 0.55, 0.53)
testing = (0.54, 0.63 , 0.65, 0.65, 0.45, 0.56)

# create plot
fig, ax = plt.subplots(figsize=(8, 6))
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, training, bar_width,
alpha=opacity,
color='b',
label='Training')

rects2 = plt.bar(index + bar_width, testing, bar_width,
alpha=opacity,
color='g',
label='Testing')

plt.xlabel('Approach')
plt.ylabel('Completeness score')
plt.title('Completeness score for approach')
plt.xticks(index + bar_width, ('FDS', 'RGB + H&E', 'CAE-MSE', 'CAE-SSIM', 'SCAE-MSE', 'SCAE-SSIM'))
plt.legend()

plt.tight_layout()
plt.savefig('results.png')
plt.show()
```
