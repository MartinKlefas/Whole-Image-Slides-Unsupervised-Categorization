```python
## Data loading, and standardizing
test_directory = "../../Data/Kather_all_test"

kather_test = []

# Test on 1000 images
for filename in os.listdir(test_directory):
    if filename.endswith('.tif'):
        image = cv2.imread(os.path.join(test_directory, filename), cv2.IMREAD_UNCHANGED)
        resized_image = cv2.resize(image, (144, 144)) 
        kather_test.append(np.asarray( resized_image, dtype="uint8" ))
        
kather_test = np.asarray(kather_test)
kather_test = kather_test.astype('float32') / 255.

## Data loading, and standardizing
train_directory = "../../Data/Kather_all_train"

kather_train = []

# Test on 1000 images
for filename in os.listdir(train_directory):
    if filename.endswith('.tif'):
        image = cv2.imread(os.path.join(train_directory, filename), cv2.IMREAD_UNCHANGED)
        resized_image = cv2.resize(image, (144, 144)) 
        kather_train.append(np.asarray( resized_image, dtype="uint8" ))
        
kather_train = np.asarray(kather_train)
kather_train = kather_train.astype('float32') / 255.

print(kather_train.shape)
kather_test.shape
```

```python
def loadAndEvaluateModel(modelpath , dataMatrix):
    mseList = []
    autoencoder = load_model(modelpath, compile=False) 
    decoded_imgs = autoencoder.predict(dataMatrix)[0]
#     print(decoded_imgs.shape[0])
    
    for i in range(decoded_imgs.shape[0]):
#         print(dataMatrix[i].shape)
#         print(decoded_imgs[i].shape)
        mseList.append(mse(dataMatrix[i] , decoded_imgs[i]))
    print("Average MSE (normalized pixels) : {}".format(sum(mseList) / len(mseList)))
```

```python
# CAER-MSE-8
loadAndEvaluateModel('../Autoencoders/SavedModels/weights/finalists/CAER_MSE_KATHER.h5' , kather_test)
loadAndEvaluateModel('../Autoencoders/SavedModels/weights/finalists/CAER_MSE_KATHER.h5' , kather_train)
```
